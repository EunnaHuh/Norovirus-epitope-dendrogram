```python
import os,math,sys, numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import pandas as pd
import seaborn as sns
```

```python
fileN='GI-1-epitopes'
DF=pd.read_csv(fileN+'.csv',index_col='ID')
display(HTML(DF.head(5).to_html()))
```

```python
rc = {'figure.figsize':(10,5),
      'axes.facecolor':'white',
      'axes.grid' : True,
      'grid.color': '1',
      'font.family':'Arial',
      'font.size' : 16}
plt.rcParams.update(rc)

g=sns.clustermap(DF,cmap='YlOrBr',linewidths=1,linecolor='gray',tree_kws=dict(linewidths=1.5),figsize=(15, 20),col_cluster=True,
                 row_cluster=False, metric="euclidean",method="ward")
g.cax.set_visible(False)
g.savefig(fileN+'.pdf', format='pdf')
```
